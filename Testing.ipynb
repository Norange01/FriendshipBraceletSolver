```python
import time
import numpy as np
from Solver import Solver
from InputGenerator import random_target_design_generator

max_elapsed_times=[]
average_elapsed_times=[]

L_bound=3
U_bound=10

for num_threads in range(L_bound,U_bound):
    print("Num threads: "+str(num_threads))
    elapsed_times=[]
    for i in range(5):
        target_design=random_target_design_generator(num_threads,4,3)
        solver= Solver(target_design)
        start_t=time.time()
        solver.solve(True)
        end_t=time.time()
        elapsed_t=end_t-start_t
        elapsed_times.append(elapsed_t)
    max_elapsed_times.append(np.max(elapsed_times))
    average_elapsed_times.append(np.mean(elapsed_times))
```

```python
import matplotlib.pyplot as plt

x=np.linspace(L_bound,U_bound,(U_bound-L_bound))

plt.plot(x,average_elapsed_times, label="Average")
plt.plot(x,max_elapsed_times, label="Maximum")

y = (2.5**x)/3000
plt.plot(x,y, '--', label="(2.5^x)/3000")

plt.title("Latency as a function of num_threads")
plt.xlabel("Number of threads")
plt.ylabel("Latency (s)")
plt.legend()
plt.yscale('log')
plt.show()
```

```python
import time

max_elapsed_times=[]
average_elapsed_times=[]

L_bound=2
U_bound=14

for num_rows in range(L_bound,U_bound, 2):
    print("Num rows: "+str(num_rows))
    elapsed_times=[]
    for i in range(10):
        target_design=random_target_design_generator(6,num_rows,4)
        solver= Solver(target_design)
        start_t=time.time()
        solver.solve(True)
        end_t=time.time()
        elapsed_t=end_t-start_t
        elapsed_times.append(elapsed_t)
    max_elapsed_times.append(np.max(elapsed_times))
    average_elapsed_times.append(np.mean(elapsed_times))
```

```python
import matplotlib.pyplot as plt

x=np.linspace(L_bound,U_bound,int((U_bound-L_bound)/2))

plt.plot(x,average_elapsed_times, label="Average")
plt.plot(x,max_elapsed_times, label="Maximum")

y = (1.7**x)/50
plt.plot(x,y, '--', label="(1.7**x)/50")

plt.title("Latency as a function of num_rows")
plt.xlabel("Number of rows")
plt.ylabel("Latency (s)")
plt.legend()
plt.yscale('log')
plt.show()
```
