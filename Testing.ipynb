```python
import time
import numpy as np

max_elapsed_times=[]
average_elapsed_times=[]

L_bound=3
U_bound=14

for num_threads in range(L_bound,U_bound):
    print("Num threads: "+str(num_threads))
    elapsed_times=[]
    for i in range(10):
        target_design=random_target_design_generator(num_threads,6,3)
        solver= Solver(target_design)
        start_t=time.time()
        solver.solve()
        end_t=time.time()
        elapsed_t=end_t-start_t
        elapsed_times.append(elapsed_t)
    max_elapsed_times.append(np.max(elapsed_times))
    average_elapsed_times.append(np.mean(elapsed_times))
```

```python
import matplotlib.pyplot as plt

plt.plot(np.linspace(L_bound,U_bound,(U_bound-L_bound)),average_elapsed_times, label="Average")
plt.plot(np.linspace(L_bound,U_bound,(U_bound-L_bound)),max_elapsed_times, label="Maximum")
plt.title("Latency as a function of num_threads")
plt.xlabel("Number of threads")
plt.ylabel("Latency (s)")
plt.legend()
plt.yscale('log')
plt.show()
```
